**Is the frequency of the word "love" in the Beatles Discography statistically significantly associated with a given year in the Beatles Discography?**


For this project, I manually built out the dataset which is comprised of all the lyrics in the Beatles *core* discography. In addition to the scope of this project, I wished to make some general graphs and potentially use this dataset for later projcts, so I went through the data by hand to clean it up. I created variables for each of the 154 songs in the dataset and created a Pandas dataframe containing the following elements:

`Song Name - Lyrics - Album Name - Year - 'Love' Count - Total Word Count - Love_Ratio`

The `Love_Ratio` is determined by dividing the `'Love' Count`, ala the frequency of the word "love" in a particular song by the total number of words in the song `(Word Count)`.

In order to count the number of times "love" was used in any particular song, I first split the words in the "Lyrics" column and then iterated through each one counting the uses of the word "love" and its different cases.

After getting a count of the word "love" - I counted the entirety of the words in any given song and created a new column in my dataframe called `Total Word Count`

The .head() of my dataframe ultimately ended up looking like this, after some simple math to get the "Love_Ratio" per song.

After setting these columns up, I plotted the "Love_Ratio" and "Year" variables using `lmplot`. The resultant graph seemed to indicate that these variables were not statistically significantly correllated, but to be certain, I used OLS to create a model of the relationship between the Love_Ratio and Year variables in the dataset. 


```python
sns.lmplot(x = 'Year', y = 'Love_Ratio', data= Beatles_Discography, ci = 95);
```

The model confirmed my hunch that these two values were not correlated, providing a very large P value of 0.453, and extremely small r^2 value of 0.004.

From the model, I can conclude there is no statistically significant association between the frequency of the word "love" and any year in the Beatles Discography. 

```python
model = ols('Love_Ratio ~ Year', data= Beatles_Discography).fit()

print(model.summary())
```
